importing one hot encoder for changing the sentences to the resutant matrix

```python
from tensorflow.keras.preprocessing.text import one_hot
```

defining sentences

```python
sent=[
    'The glass of milk',
    'the glass of water',
    'the cup of tea',
    'i am a good boy',
    'I am a good Coder',
    'I AM GOOD',
    'understanding what I say?'
]
```

defining vocabulary size for the sentences

```python
vocab_size=10000
```

nowrepresenting the sentences as one hot encoding

```python
onehot_repr=[one_hot(words,vocab_size) for words in sent]
print(onehot_repr)
```

in above we can observe that same word have same index

now representation as word embedding

```python
from tensorflow.keras.layers import Embedding
```

```python
from tensorflow.keras.models import Sequential
```

```python
#for padding purpose
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
import numpy as np
```

```python
#defining sentence length by which the length will not exceed
sen_len=8
```

```python
#we are using pre as it will give zero before all if we use post the 0 at last
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sen_len)
print(embedded_docs)
```

defining dimensions which we want to add

```python
dim=10
```

defining models

```python
model=Sequential()
model.add(Embedding(vocab_size,dim,input_length=sen_len))
```

```python
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```
